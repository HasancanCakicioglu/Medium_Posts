# **Bernoulli Distribution**

A probability distribution that models the outcomes of a single binary experiment, such as success or failure.

For example, let's consider flipping a fair coin. We can define success as getting heads (H) and failure as getting tails (T). In this case, p = 0.5 because the probability of getting heads is 0.5.

```python
import numpy as np
import matplotlib.pyplot as plt

# Set the probability of getting heads
p = 0.5

# Simulate flipping a coin 100 times
coin_flips = np.random.choice([0, 1], size=100, p=[1-p, p])

# Count the number of heads and tails
heads_count = np.count_nonzero(coin_flips == 1)
tails_count = np.count_nonzero(coin_flips == 0)

# Plot the results
labels = ['Heads', 'Tails']
counts = [heads_count, tails_count]

plt.bar(labels, counts)
plt.xlabel('Outcome')
plt.ylabel('Count')
plt.title('Bernoulli Distribution - Coin Flips')
plt.show()

print(heads_count)
print(tails_count)
```

# **Binomial Distribution**

The Binomial Distribution is used to model the number of successes in a fixed number of independent Bernoulli trials, where each trial has a constant probability of success. It is commonly used to analyze outcomes with two possible outcomes, such as flipping a coin or the success/failure of a product.

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import binom

n = 10
p = 0.5

# Calculate the probabilities for all possible outcomes
k_values = np.arange(0, n+1)
probabilities = binom.pmf(k_values, n, p)

# Plot the probability mass function
plt.bar(k_values, probabilities)
plt.xlabel('Number of Successes')
plt.ylabel('Probability')
plt.title('Binomial Distribution')
plt.show()
```

# **Poisson Distribution**

The Poisson Distribution is used to model the number of events that occur in a fixed interval of time or space, where the events occur with a constant average rate and independently of the time since the last event. It is commonly used to analyze rare events that occur randomly, such as the number of emails received per hour or the number of car accidents in a day.

For example, let's consider the scenario of counting the number of customer arrivals at a store in an hour, where on average, 5 customers arrive per hour.

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import poisson

u = 5

# Calculate the probabilities for all possible number of events
x_values = np.arange(0, 15)
probabilities = poisson.pmf(x_values, u)

# Plot the probability mass function
plt.bar(x_values, probabilities)
plt.xlabel('Number of Events')
plt.ylabel('Probability')
plt.title('Poisson Distribution')
plt.show()
```

# **Hypergeometric Distribution**

Hypergeometric Distribution is a probability distribution that models the number of successful outcomes in a finite population, without replacement. It is often used when sampling from a population with distinct categories or subgroups.


For example,In a population of 100 employees, where 60 are male and 40 are female, if we randomly select 10 employees, the Hypergeometric Distribution can be used to calculate the probability of obtaining a specific number of male or female employees in the sample.

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import hypergeom

# Define the population and sample size
N = 100
m = 60
n = 10

# Create a hypergeometric distribution object
hypergeom_dist = hypergeom(N, m, n)

# Generate the probability mass function
x = np.arange(0, n+1)
pmf = hypergeom_dist.pmf(x)

# Plot the probability mass function
plt.bar(x, pmf)
plt.xlabel('Number of Successes')
plt.ylabel('Probability')
plt.title('Hypergeometric Distribution')
plt.show()
```

# **Negative Binomial Distribution**

Negative Binomial Distribution is a probability distribution that models the number of failures that occur before a specified number of successes in a sequence of independent Bernoulli trials.

In a basketball game, the Negative Binomial Distribution can be used to calculate the probability of the number of missed shots until a player makes 5 successful shots.

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import nbinom

# Define the number of successes and the probability of success
r = 5
p = 0.3

# Create a negative binomial distribution object
nbinom_dist = nbinom(r, p)

# Generate the probability mass function
x = np.arange(0, 20)
pmf = nbinom_dist.pmf(x)

# Plot the probability mass function
plt.bar(x, pmf)
plt.xlabel('Number of Failures')
plt.ylabel('Probability')
plt.title('Negative Binomial Distribution')
plt.show()
```

# **Normal (Gaussian) Distribution**

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import norm

# Parameters
mean_weight = 70      # Mean weight in kilograms
std_dev = 8          # Standard deviation in kilograms

# Generate data points
x = np.linspace(mean_weight - 3 * std_dev, mean_weight + 3 * std_dev, 1000)

# Calculate the probability density function (PDF)
pdf = norm.pdf(x, mean_weight, std_dev)

# Plot the PDF
plt.plot(x, pdf)
plt.xlabel('Weight (kg)')
plt.ylabel('Probability Density')
plt.title('Normal Distribution')
plt.grid(True)
plt.show()

```

```python
import numpy as np
import matplotlib.pyplot as plt

# Parameters
a = 8.0  # Lower bound of the time window (e.g., 8:00 AM)
b = 9.0  # Upper bound of the time window (e.g., 9:00 AM)

# Generate data points
x = np.random.uniform(a, b, 1000)

# Plot the probability density function
plt.plot(np.sort(x), np.full_like(x, 1/(b-a)), '-')
plt.xlabel('Time (in hours)')
plt.ylabel('Probability Density')
plt.title('Continuous Uniform Distribution - Bus Arrival Times')
plt.grid(True)
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

# Parameter
lambda_ = 0.1  # Average arrival rate of customers per minute

# Generate data points
x = np.linspace(0, 30, 1000)  # Generate 1000 values for the x-axis (waiting time)

# Compute the probability density function (PDF)
pdf = lambda_ * np.exp(-lambda_ * x)  # Exponential Distribution formula

# Plot the probability density function
plt.plot(x, pdf)
plt.xlabel('Waiting Time')
plt.ylabel('Probability Density')
plt.title('Exponential Distribution - Customer Arrival Times')
plt.grid(True)
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import geom

# Define the probability of success
p = 0.7  # Probability of making a free throw in a single trial

# Create a geometric distribution object
geom_dist = geom(p)

# Generate the probability mass function
k = np.arange(1, 11)  # Number of trials
pmf = geom_dist.pmf(k)

# Plot the probability mass function
plt.bar(k, pmf)
plt.xlabel('Number of Trials (k)')
plt.ylabel('Probability')
plt.title('Geometric Distribution')
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import lognorm

# Parameters for the Log-Normal Distribution
mu = 0.05  # Mean of the logarithm (expected return)
sigma = 0.2  # Standard deviation of the logarithm (volatility)

# Create a log-normal distribution object
lognorm_dist = lognorm(s=sigma, scale=np.exp(mu))

# Generate random samples from the distribution
samples = lognorm_dist.rvs(size=1000)

# Generate range of x values
x = np.linspace(0, np.max(samples), 100)

# Compute the probability density function (PDF) for the x values
pdf = lognorm_dist.pdf(x)

# Plot the continuous line representing the PDF
plt.plot(x, pdf, color='red')

plt.xlabel('Price')
plt.ylabel('Probability Density')
plt.title('Log-Normal Distribution: Stock Price')
plt.grid(True)
plt.show()
```
